```python
import pandas as pd
import numpy as np
import sqlite3
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
conn=sqlite3.connect('car_sales.db')
```

```python
table=pd.read_sql("select name from sqlite_master where type ='table'",conn)
table
```

```python
df=pd.read_sql("select * from Car",conn )
```

```python
categorial_cols=['Customer Name','Company','Model','Transmission','Color']

plt.figure(figsize=(12,10))
for i,col in enumerate(categorial_cols):
    plt.subplot(2,3,i+1)
    sns.countplot(y=df[col],order =df[col].value_counts().index[:10])
    plt.title(f"count plot of {col}")
    plt.tight_layout()
plt.show()
```
