```python
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///car_sales.db')
```

```python
def ingest_db(df, table_name, engine):
    df.to_sql(table_name, con=engine, if_exists='replace', index=False)

folder_path = r'C:\Users\46575\python project\car sales'

for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)  # Full path
        df = pd.read_csv(file_path)

        table_name = os.path.splitext(file)[0]  # Correct table name
        ingest_db(df, table_name, engine)
        print(f"Inserted {file} into table {table_name}")

print("✅ All CSV files have been loaded into the database!")
```
